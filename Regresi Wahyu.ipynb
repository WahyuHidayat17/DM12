```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
#memanggil dataset
dataset = pd.read_excel('datasetTokoA.xlsx')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
dataset.keys()
```

```python
dataset.shape
```

```python
#menampilkan isi sebagian dataset
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
#split dataset menjadi training set dan testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
#Melakukan fitting simple linear regression pada training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#Memprediksi hasil test-set
y_pred = regressor.predict(x_test)
```

```python
plt.scatter(dataku.beli, dataku.pendapatan)
plt.xlabel("Pendapan Rata-Rata")
plt.ylabel("Harga Beli")
plt.title("Grafik Pendapan Rata-Rata vs Harga Beli")
plt.show
```

```python
#Visualisasi Hasil Prediksi Pada Training-set
#Ukuran plot
plt.figure(figsize=(5,8))


#Garis merah dalah hasil prediksi dari machine learning
plt.plot(x_train, regressor.predict(x_train), color ='yellow')


#Biru adalah observasi
#//plt.Scatter(x_train, y_train, color = 'red')

#Memberi judul dan label
plt.title('Keuntungan Toko')
plt.xlabel('Harga Beli')
plt.ylabel('Pendapatan')

#menampilkan plot
plt.show()
```
